```python
def costFunction(x):
    c,e,g = x
    totalCost = 0
    data = {}

    indicies = ['136792']#, "136805", "136782", "136759", "136683", "136764", "136778", "136309", "136718", "136315", "136730", "132231", "136723", "136682"]
    for index  in indicies:
        #print()
        #print(index)
        for samples in nSamples:
            with open(f'PTSZ-instancje/{index}/in{index}_{samples}.txt', "r") as instance:
                d = instance.read().split('\n')
                n = int(d[0])
                machines = {}
                for name, speed in zip(machineNames, d[1].split(' ')):
                    machines[name] = {'speed':speed, 'readyTime':0, 'gain':0}
                for i,line in enumerate(d[2:]):
                    if line == '':
                        continue
                    p,r = line.split(' ')
                    p,r = int(p), int(r)
                    data[i+1] = {"p": p, "r": r, "c": 0, "penalty": 0 }
                sortedData = sorted(data.items(), key=lambda y: float(y[1]['r']))
                for key, value in sortedData:
                    data[key]: value
                dataCopy = copy.deepcopy(data)

            calculatedWeight = 0
            timeCounter = 0
            outSchedule = {}
            for name in machineNames:
                outSchedule[name] = []

            start = time.time()
            while len(data) > 0:
                readyToGo = {}
                for key, task in data.items():
                    if task['r'] <= timeCounter:
                        #endTime = timeCounter+task['p']
                        #penaltyEndTask = ( endTime if endTime > 0 else 0 ) * 1
                        penaltyPersistance = task['p']
                        task['penalty'] = penaltyPersistance
                        readyToGo[key] = task

                if len(readyToGo) == 0: # currently there is no task to go
                    #print("upsi readyToGo")
                    timeCounter = min(data.items(), key=lambda x: x[1]['r'])[1]['r']
                    continue
                #print(f"readyToGo: {readyToGo}")

                # select task
                readyToGoSorted = sorted(readyToGo.items(), key=lambda y: (y[1]['penalty']), reverse=False)

                selectedTask = readyToGoSorted[0]

                # select machine
                readyMachines = [d for d in zip(machines.keys(), machines.values() ) if d[1]['readyTime'] <= timeCounter]
                if len(readyMachines) == 0: # currently there is no free machine
                    #print("upsi readyMachines")
                    timeCounter = min(machines.items(), key=lambda x: x[1]['readyTime'])[1]['readyTime']
                    continue

                readyTime = []
                minNextMachine = np.inf
                speedNextMachine = 0
                for key, m in zip(machines.keys(),machines.values() ):
                    readyTime.append(machines[key]["readyTime"])
                readyTime = np.array(readyTime)
                try:
                    minNextMachine = np.min(readyTime[readyTime > timeCounter])
                    ind = np.argwhere(readyTime==minNextMachine)[0][0]
                    speedNextMachine = machines[f'm{ind}']['speed']
                except:
                    pass

                for name, m in readyMachines:
                    speed = machines[name]['speed']

                    if minNextMachine != np.inf:
                        machines[name]['gain'] = c*float(speed) + e*minNextMachine + g*float(speedNextMachine)
                    else:
                        machines[name]['gain'] = c*float(speed)

                selectedMachine = max(readyMachines, key=lambda y: (y[1]['gain']))

                if machines[selectedMachine[0]]['readyTime'] > timeCounter: #selectedMachine is busy then wait
                    print("WHAT THE???")
                    timeCounter = machines[selectedMachine[0]]['readyTime']

       #         print(f"timeCounter: {timeCounter}")
           #     print(f"selectedTask: {selectedTask}")

                # remove task from list active tasks, add time, append outschedule
                data.pop(selectedTask[0])

        #        print(f"selectedMachine: {selectedMachine}")
                #timeCounter += 1
                machines[selectedMachine[0]]['readyTime'] = timeCounter + (selectedTask[1]['p'] / float(selectedMachine[1]['speed']) )
         #       print(machines)

          #      print()

                outSchedule[selectedMachine[0]].append(selectedTask[0])    

            end = time.time()
            #print(end-start)

            cost = calculateCost(dataCopy, outSchedule, machines)
            cost /= samples
            totalCost += cost
    print(f"{totalCost}   [{c},{e},{g}],")
    return totalCost
```
