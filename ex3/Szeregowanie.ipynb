```python
def calculateCost(data, taskQueue):
    calculatedWeight = 0
    m1Ready = 0
    m2Ready = 0
    m3Ready = 0
    totalW = 0
    for task in taskQueue:        
        m1Ready += data[task]["p1"]
       
        m2Ready = max(m1Ready, m2Ready)
        m2Ready += data[task]["p2"]
        
        m3Ready = max(m2Ready, m3Ready)
        m3Ready += data[task]["p3"]
        
            
        D = max(m3Ready - data[task]['d'], 0)
        calculatedWeight += D * data[task]["w"]
        totalW += data[task]['w']
    #print(calculatedWeight)
    return calculatedWeight / totalW
```
